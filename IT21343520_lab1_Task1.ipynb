#4. Correlation Analysis with Heatmaps

```python
data = pd.DataFrame({
    'HP': [45, 60, 80, 39, 58, 78, 44, 50, 65],
    'Attack': [49, 62, 82, 52, 64, 84, 48, 52, 80],
    'Defense': [49, 63, 83, 43, 58, 78, 65, 50, 55],
    'Sp. Atk': [65, 80, 100, 60, 80, 100, 50, 65, 90],
    'Sp. Def': [65, 80, 100, 50, 60, 85, 64, 64, 80],
    'Speed': [45, 60, 80, 65, 70, 100, 43, 50, 85]
})

# Calculating Pearson and Spearman correlations
pearson_corr = data.corr(method='pearson')
spearman_corr = data.corr(method='spearman')

# Plotting heatmaps
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', ax=ax[0])
ax[0].set_title('Pearson Correlation Heatmap')
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', ax=ax[1])
ax[1].set_title('Spearman Correlation Heatmap')
plt.show()
```
